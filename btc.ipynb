```python
import pandas as pd
data = pd.read_csv('data.csv')
data
```

```python
date = data['Date']
date
```

```python
type(date)
```

```python
date
```

```python
lst = []
for i in date:
    x = i.replace('-','/')
    lst.append(x)
lst
```

```python
maindata = []
for i in lst:
    x = i.split('/')
    z = x[::-1]
    maindata.append(z)
maindata
```

```python
ref = maindata
lstt = []
for i in ref:
    x = i[0] + '-' +  i[2] + '-' + i[1] + " " 
    lstt.append(x)
```

```python
lstt = lstt + data['Time'] + ':00-05:00'
# lstt
lstt
```

```python
exchange = ccxt.binance()
import time
priceLst = []
for i in lstt:
    timestamp = int(datetime.datetime.strptime(i, "%Y-%m-%d %H:%M:%S%z").timestamp() * 1000)
    response = exchange.fetch_ohlcv('BTC/USDT', '1m', timestamp, 1)
    priceLst.append(response)
    time.sleep(0.5)
```

```python
priceLst
```

```python
mylist = []
for i in priceLst:
    x = i[0]
    mylist.append(x)
```

```python
df = pd.DataFrame(mylist)
df
```

```python
df = df[[1, 4]]
df
df.rename(columns={1: 'OpeningPriceAtMinute', 4: 'ClosingPriceAtMinute'}, inplace=True)
df = df.to_csv('price.csv')
# df
```
